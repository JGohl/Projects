```python
# Create code to answer each of the following questions.
# Hint: You will need multiple target URLs and multiple API requests.

# Dependencies
import requests
import json
from pprint import pprint

# Retrieve Google API key from config.py
from config import gkey
```

```python
# 1. What are the geocoordinates (latitude/longitude) of Seattle, Washington?
# Run a request to endpoint and convert result to json
# Target city
target_city = "Seattle, Washington"

# Build the endpoint URL
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_city, gkey)

geo_data = requests.get(target_url).json()

# Print the json
# pprint(geo_data)

# Extract latitude and longitude
lat_seattle = geo_data["results"][0]["geometry"]["location"]["lat"]
lng_seattle = geo_data["results"][0]["geometry"]["location"]["lng"]

# Print the latitude and longitude
print('''
    City: {0}
    Latitude: {1}
    Longitude: {2}
    '''.format(target_city, lat_seattle, lng_seattle))
```

```python
# 3. Find the name andaddress of a bike store in Seattle, Washington.
#    Hint: See https://developers.google.com/places/web-service/supported_types

# geocoordinates
target_coordinates_seattle = "47.6062095, -122.3320708"
target_search_seattle = "bike shops"
target_radius_seattle = 80000
target_type_seattle = "bicycle_store"

# set up a parameters dictionary
params = {
    "location": target_coordinates_seattle,
    "keyword": target_search_seattle,
    "radius": target_radius_seattle,
    "type": target_type_seattle,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
places_data = response.json()

#pprint(places_data)
print(places_data["results"][1]["name"])
print(places_data["results"][1]["vicinity"])
```
