```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Import API key
from config import gkey
```

```python
# Import cities file into the cities_pd DataFrame
file = "../Resources/cities.csv"
cities_pd = pd.read_csv(file)
cities_df = pd.DataFrame(cities_pd)
cities_df.head()
```

```python
# Add columns for lat, lng, airport name, airport address, airport rating.
# Hint: Be sure to specify "" as the initial value for each column.
cities_df['Lat'] = ''
cities_df['Lng'] = ''
cities_df['Aiport Name'] = ''
cities_df['Airport Address'] = ''
cities_df['Airport Rating'] = ''
cities_df.head()
```

```python
# Loop through the cities_pd and get the lat/long for each city

# Build the endpoint URL

for index, row in cities_df.iterrows():
    
    target_city = row[0]
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
        'address={0}&key={1}').format(target_city, gkey)
    geo_data = requests.get(target_url).json()
    
    cities_df.loc[index, 'Lat'] = geo_data["results"][0]["geometry"]["location"]["lat"]
    cities_df.loc[index, 'Lng'] = geo_data["results"][0]["geometry"]["location"]["lng"]

# Hint: `requests.get(target_url).json()`
# Hint: `for index, row in cities_pd.iterrows():`
# Hint: `cities_pd.loc`

cities_df.head()
```
